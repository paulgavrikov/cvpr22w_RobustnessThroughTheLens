```python
dataset_path = "/data/output/20220226_robustness/dataset.h5"
df_meta = pd.read_hdf(dataset_path, "meta")
df_meta.Robust = df_meta.Robust.apply(bool)
```

```python
df_meta.Paper.unique()
```

```python
import re
import urllib.request as libreq

ids = []

paper_map = {}

for s in df_meta.Paper.unique():
    if type(s) is str:
        match = re.findall(r"arxiv.*([0-9]{4}.[0-9]{4,5})", s)
        if len(match) > 0:
            idd = match[0]
            ids.append(idd)
            
for idd in set(ids):
    url = f"https://arxiv.org/bibtex/{idd}"
    with libreq.urlopen(url) as data:
        citation = data.read().decode("utf-8")
        print(citation)
        paper_map[idd] = citation
```
